```python
## 실습 7 wc
f = open('file.txt','rt',encoding="UTF-8")
line_count, word_count = 0, 0
for line in f:
    if not line:
        break
    line_count+=1
    word_list = line.split()
    word_count += len(word_list)
f.close()
print('line_count : ',line_count)
print('word_count : ',word_count)
```
