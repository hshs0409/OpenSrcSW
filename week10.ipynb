```python
first_list = input().split(' ')
temp_list = []
for s_idx, str in (enumerate(reversed(first_list))):
    temp_str = ""
    for idx, val in enumerate(str):
        temp_str += chr(ord(val) + 1)
    temp_list.append(temp_str) 
print(' '.join(temp_list))

second_list = input().split(' ')
temp_list = []
for s_idx, str in (enumerate(reversed(second_list))):
    temp_str = ""
    for idx, val in enumerate(str):
        temp_str += chr(ord(val) - 1)
    temp_list.append(temp_str) 
print(' '.join(temp_list))


```

```python
phone_book = dict()
for _ in range(10):
    name, phone = input("이름 전화번호 입력 : ").split()
    phone_book[name] = phone

def serach_by_name(name):
    if phone_book.get(name):
        print(name, phone_book.get(name))
    else:
        print('해당 이름이 존재하지 않습니다.')

def search_by_pnum(pnum):
    for name, phone in phone_book.items():
        if int(phone) == pnum:
            print(f"{name} : {pnum}")
            return
    print('해당 번호가 존재하지 않습니다.')

while True:
    menu_choice = int(input("1 : 이름으로 찾기 2: 번호로 찾기 3: 종료 "))
    if menu_choice == 1:
        serach_name = input("찾을 이름을 입력하세요 : ")
        serach_by_name(serach_name)
    elif menu_choice == 2:
        serach_pnum = int(input("찾을 번호을 입력하세요 : "))
        search_by_pnum(serach_pnum)
    elif menu_choice == 3:
        break
```
