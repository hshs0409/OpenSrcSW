```python
## 일교차가 가장 큰 날, 최저기온이 가장 낮은 날,최고기온이 가장 높은 날 출력
import csv

f = open('seoul.csv', 'r', encoding='cp949')
data = csv.reader(f)
max_day_list = [-999, 0, 999, 0, -999, 0]
header = next(data)
for row in data:
    if row[0] == '':
        break
    if max(float(row[4]) - float(row[3]),
           max_day_list[0]) == float(row[4]) - float(row[3]):
        max_day_list[0], max_day_list[1] = float(row[4]) - float(
            row[3]), row[0]
    if min(float(row[3]), max_day_list[2]) == float(row[3]):
        max_day_list[2], max_day_list[3] = float(row[3]), row[0]
    if max(float(row[4]), max_day_list[4]) == float(row[4]):
        max_day_list[4], max_day_list[5] = float(row[4]), row[0]
f.close()
print('일교차가 가장 큰 날')
print(max_day_list[1])
print('최저기온이 가장 낮은 날')
print(max_day_list[3])
print('최고기온이 가장 높은 날')
print(max_day_list[5])
```

```python
## 노선별 이용객 가장 많은 3개역 가장 적은 3개역 출력
import csv
from collections import Counter

f = open('subwayfee.csv', 'r', encoding='cp949')
data = csv.reader(f)
header = next(data)

print(header)
data = list(data)
data.sort(key=lambda x: (x[1],int(x[4].replace(',',''))+int(x[5].replace(',',''))))
f.close()

way_list=[]
for way in data:
    way_list.append(way[1])
way_list=Counter(way_list)
way_len = 0
for key in way_list:
    print('--이용객이 가장 적은 역 순위--')
    print(data[way_len+0][1] ,data[way_len+0][3] ,int(data[way_len+0][4].replace(',',''))+int(data[way_len+0][5].replace(',','')))
    print(data[way_len+1][1] ,data[way_len+1][3] ,int(data[way_len+1][4].replace(',',''))+int(data[way_len+1][5].replace(',','')))
    print(data[way_len+2][1] ,data[way_len+2][3] ,int(data[way_len+2][4].replace(',',''))+int(data[way_len+2][5].replace(',','')))
    way_len += way_list[key]
    
    print('--이용객이 가장 많은 역 순위--')
    print(data[way_len-1][1] ,data[way_len-1][3] ,int(data[way_len-1][4].replace(',',''))+int(data[way_len-1][5].replace(',','')))
    print(data[way_len-2][1] ,data[way_len-2][3] ,int(data[way_len-2][4].replace(',',''))+int(data[way_len-2][5].replace(',','')))
    print(data[way_len-3][1] ,data[way_len-3][3] ,int(data[way_len-3][4].replace(',',''))+int(data[way_len-3][5].replace(',','')))
    print()
    
    
```
