```python
import matplotlib.pyplot as plt
import csv
f = open('birthweather.csv', 'r', encoding='cp949')
data = csv.reader(f)
header = next(data)
h_temp , m_temp =[], []
for day in list(data):
    if '04-09' in day[0]:
        m_temp.append(float(day[3]))
        h_temp.append(float(day[4]))

    
plt.title('2011~2020 0409 highest, minimum temperature')
plt.plot(h_temp,'pink')
plt.plot(m_temp,'skyblue')
plt.show()
```

```python
import matplotlib.pyplot as plt
import csv
f = open('birthweather.csv', 'r', encoding='cp949')
data = csv.reader(f)
header = next(data)
jan_h_temp,aug_h_temp = [],[]
for day in list(data):
    if '-01-' in day[0]:
        jan_h_temp.append(float(day[-1]))
    if '-08-' in day[0]:
        aug_h_temp.append(float(day[-1]))
    
plt.title('2020-01, 2020-08 highest temperature histo')
plt.hist(aug_h_temp,bins=100,color='red',label='Aug')
plt.hist(jan_h_temp,bins=100,color='blue',label='Jan')
plt.legend()
plt.show()
```
